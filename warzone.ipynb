## Popularity based recommendation

```python
import pandas as pd
import numpy as np
```

### import dataset

```python
song_df = pd.read_csv('dataset/song_dataset.csv', dtype={"year": "string"})
song_df.head(10)
```

```python
user_df = pd.read_csv('dataset/user_dataset.csv')
user_df.head(10)
```

```python
song_df.drop_duplicates(subset='song_id', keep = 'first', inplace = True)
song_df.head(10)
len(song_df)
```

```python
con_df = pd.merge(user_df, song_df, on='song_id', how='left')
con_df.head(10)
```

```python
con_df['song'] = con_df['title'] + ' - ' + con_df['artist_name']
con_df.head(10)
```

```python
song_grouped = con_df.groupby(['song']).agg({'listen_count' : 'count'}).reset_index()
song_grouped.head(10)
```

```python
song_grouped.sort_values(by=['listen_count'], ascending=False).head(10)
```

```python
grouped_sum = song_grouped['listen_count'].sum()
song_grouped['percentage'] = song_grouped['listen_count'] / grouped_sum * 100
song_grouped = song_grouped.sort_values(['listen_count', 'song'], ascending=[0, 1])
song_grouped.head(10)
```

```python
song_grouped['rank'] = song_grouped['listen_count'].rank(ascending=0, method='first')
song_grouped.head(10)
```

```python
user1 = user_df['user_id'][5]
user_data = user_df[user_df['user_id'] == user1]
user_data = user_data.sort_values('listen_count', ascending=False)
list(user_data['song_id'].head(10))
```
